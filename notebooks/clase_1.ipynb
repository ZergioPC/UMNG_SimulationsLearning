## Metodo de Euler

```python
def fwd_euler(f, a, b, y0, h):
    n = round((b-a)/h)
    h = (b-a)/n
    #t = [0]*(n+1)
    t = [0 for i in range(n+1)]
    y = [0]*(n+1)
    t[0] = a
    y[0] = y0
    for k in range(0, n):
        y[k+1] = y[k] + h*f(t[k],y[k])
        t[k+1] = t[k] + h
    return t, y
```

## Ecuaciones

La ecuaci칩n diferencial utilizada es:

$
\frac{dy}{dt} = -2 t y
$

```python
def edo1(t,y):
    salida=-2*t*y
    return salida
```

La soluci칩n exacta de la ecuaci칩n diferencial es:

$
y(t) = e^{-t^2}
$

```python
def sol1(t):
    y=np.exp(-t**2)
    return y
```

## Plot Gr치ficas

```python
#t=np.arange(0,5,0.3)
t2=np.linspace(0,4,101)
yexacta= sol1(t2)

taprox, yaprox=fwd_euler(edo1, 0, 4, 1, 0.3)

plt.plot(t2,yexacta)
plt.plot(taprox,yaprox,'*-')

#plt.title("hola")
plt.show()
```
